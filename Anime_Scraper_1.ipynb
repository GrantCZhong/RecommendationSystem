```python
import pandas as pd

wd = webdriver.Chrome('chromedriver', options=chrome_options)
driver = webdriver.Chrome('chromedriver', options=chrome_options)
```

```python
## Extract Anime Name + Ratings ##
counter = 0
base_url = 'https://myanimelist.net/topanime.php?type=tv&limit='

df = pd.DataFrame(columns=['Title','Rating'])
df_count = 0
while (counter <=450):
  driver.get(base_url+str(counter))
  for i in range(50):
    print(df_count)
    xpath_name_0 = '/html/body/div[1]/div[2]/div[3]/div[2]/div[4]/table/tbody/tr['
    xpath_rating_0 = '/html/body/div[1]/div[2]/div[3]/div[2]/div[4]/table/tbody/tr['
    xpath_name_1 = ']/td[2]/div/div[2]/a[1]'
    xpath_rating_1 = ']/td[3]/div/span'
    xpath_name = xpath_name_0 + str(i+2) + xpath_name_1
    xpath_rating = xpath_rating_0 + str(i+2) + xpath_rating_1
    name = driver.find_elements_by_xpath(xpath_name)[0].text
    rating = driver.find_elements_by_xpath(xpath_rating)[0].text
    df.loc[df_count,'Title'] = name
    df.loc[df_count,'Rating'] = rating
    df_count+=1

  counter+=50


    

```
